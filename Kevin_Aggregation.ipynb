```python
import pandas as pd
from datetime import datetime
%pylab inline
```

```python
df = pd.read_csv("merged_bike_data.csv",dtype=unicode)
df['duration_minutes'] = df['duration_minutes'].astype(int)
df['TempAvgF'] = df['TempAvgF'].astype(int)
df['month'] = df['month'].astype(float)
df['year'] = df['year'].astype(float)
```

```python
df.columns.values
```

```python
df['start_time'] = pd.to_datetime(df['start_time'])
df.set_index('start_time', inplace=True)
df[-5:]
```

```python
df['year'] = df.index.year
df[df['year']==2016]['year']
```

```python
df = df[df['month'].notna()]

duration_weekly = df['duration_minutes'].resample('W', kind='period').agg(['mean','median'])
duration_monthly = df['duration_minutes'].resample('M', kind='period').agg(['mean','median'])
```

```python
duration_weekly.plot(title='Duration of Rides By Week')
ylabel('Duration of Rides (Minutes)')
xlabel('Month')
show()
```

```python
duration_monthly.plot(title='Duration of Rides By Month')
ylabel('Duration of Rides (Minutes)')
xlabel('Month')
show()
```
